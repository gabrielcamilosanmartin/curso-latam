## Modelación descriptiva

Como se dijo en un inicio se trata de un problema de clasificacion por lo que utilizaremos un modelo de clasificacion con la funcion logit de la libreria statsmodels.formula.api

$$\log\left(\frac{p({ income = 1})}{1 - p({ income = 1})}\right)={{\beta }_{0} + \sum_{j=0}^N {\beta }_{j} \cdot {X}}$$

Iniciamos incluyendo todas las variables utilizando un nivel de confianza del 95%
