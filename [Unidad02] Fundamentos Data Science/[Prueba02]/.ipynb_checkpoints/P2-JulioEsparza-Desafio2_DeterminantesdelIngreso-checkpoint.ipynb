## Hito 1: Aspectos Computacionales

```python
#1.- Librerías y módulos a implementar

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math


import seaborn as sns
# scipy stats para simular
import scipy.stats as stats
# statsmodels para modelación
import statsmodels.api as sm
import statsmodels.formula.api as smf
import missingno as msgn

plt.style.use('seaborn')

import warnings
warnings.filterwarnings("ignore")


#Modelo predictivo

from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

##Estandarizamos modelo prueba

from sklearn.preprocessing import StandardScaler
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score



#Métricas evaluación
from sklearn.metrics import classification_report
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score


#2.- Funciones generadas y su objetivo.

def inspeccion_covarianza(df,vector):
    columns = df.columns
    attr_name = []
    pearson_r = []
    abs_pearson_r = []
    for col in columns:
        if col != vector:
            attr_name.append(col)
            pearson_r.append(df[col].corr(df[vector]))
            abs_pearson_r.append(abs(df[col].corr(df[vector])))
            features = pd.DataFrame({'attribute': attr_name,
                                     'corr':pearson_r,
                                     'abs_corr':abs_pearson_r})
    features = features.set_index('attribute')
    features.sort_values(by=['abs_corr'], ascending=False)
    return features


def inspeccion_valores(df):
    for i in df:
        if len(df[i].value_counts()) > 2:
            print(df[i].describe(), "\n")
        else:
            print(df[i].value_counts('%'), "\n")
    return

def inspeccion_visual(df):
    sns.set(style='whitegrid', rc={"grid.linewidth": 1.1})
    sns.set_context("paper", font_scale=2)  
    for n, i in enumerate(df,1):
        plt.figure(figsize=(10,100))
        plt.subplot(25, 1, n)
        if len(df[i].value_counts()) > 2:
            sns.distplot(df[i])
            plt.title(i)
            plt.xlabel("")        
        else:
            sns.countplot(y=df[i])
            plt.title(i)
            plt.xlabel("")
            plt.tight_layout()
        return

def define_nan(df):
    NA = float('nan')
    for i in df.columns:
        for j in range(df.shape[0]):
            if df[i][j] == 'NaN':
                df[i][j].replace(['NaN'],[NA])
    return
    
```

### Limpieza y Recodificación de Datos

```python
#Se ingresa base de datos a trabajar.
df = pd.read_csv('income-db.csv')
#Reemplaza nombres de columnas según lo requerido en instrucciones.

df.rename(columns={'workclass':'workclass_recod',
                          'education':'educ_recod','marital-status':'civstatus',
                           'occupation':'collars','native-country':'region'}, 
                 inplace=True)

#Por cada columna que debe ser intervenida, se agrupa información según lo requerido en instrucciones 
#para posteriormente crear una nueva columna que numera la información para poder procesarla dentro del modelo a utilizar.

NA = float('nan')

#Para variable 'workclass_recod'

df['workclass_recod'] = df['workclass_recod'].replace(
    
    ['?',  'Federal-gov','State-level-gov','State-gov',      'Local-gov',      'Self-employed','Self-emp-inc',  'Self-emp-not-inc','Unemployed', 'Never-worked', 'Without-pay','Private'],
    [NA,'federal-gov','state-level-gov','state-level-gov','state-level-gov','self-employed','self-employed', 'self-employed',   'unemployed', 'unemployed',   'unemployed', 'private'])

# Se crean variables dummy para 'workclass_recod'

df['federal-gov_dummy'] = np.where(df['workclass_recod'] == 'federal-gov',1.0,0.0)
df['state-level-gov_dummy'] = np.where(df['workclass_recod'] == 'state-level-gov',1.0,0.0)
df['self-employed_dummy'] = np.where(df['workclass_recod'] == 'self-employed',1.0,0.0)
df['private_dummy'] = np.where(df['workclass_recod'] == 'private',1,0)
df['unemployed_dummy'] = np.where(df['workclass_recod'] == 'unemployed',1.0,0.0)


#Para variable 'educ_recod'

df['educ_recod'] = df['educ_recod'].replace(

['?',  'Preschool',      '1st-4th',     '5th-6th',           '7th-8th',    '9th',        '10th',       '11th',       '12th',       'HS-grad',    'Assoc-voc','Assoc-acdm','Some-college','Bachelors', 'Masters',   'Prof-school','Doctorate' ],
[NA,'preschool','elementary-school', 'elementary-school', 'high-school','high-school','high-school','high-school','high-school','high-school','college',  'college',   'college',     'university','university','university', 'university'])
    

# Se crean variables dummy para 'educ_recod'

df['preschool_dummy'] = np.where(df['educ_recod'] == 'preschool',1.0,0.0)
df['elementary-school_dummy'] = np.where(df['educ_recod'] == 'elementary-school',1.0,0.0)
df['high-school_dummy'] = np.where(df['educ_recod'] == 'elementary-school',1.0,0.0)
df['college_dummy'] = np.where(df['educ_recod'] == 'college',1.0,0.0)
df['university_dummy'] = np.where(df['educ_recod'] == 'university',1.0,0.0)


#Para variable 'civstatus'
    
df['civstatus'] = df['civstatus'].replace(

['?', 'Married-civ-spouse','Married-spouse-absent','Married-AF-spouse','Divorced','Separated','Widowed','Never-married'],
[NA,'married',           'married',              'married',        'divorced','separated','widowed', 'never-married'])


# Se crean variables dummy para 'civstatus'


df['married_dummy'] = np.where(df['civstatus'] == 'married',1.0,0.0)
df['divorced_dummy'] = np.where(df['civstatus'] == 'divorced',1.0,0.0)
df['separated_dummy'] = np.where(df['civstatus'] == 'separated',1.0,0.0)
df['widowed_dummy'] = np.where(df['civstatus'] == 'widowed',1.0,0.0)
df['never-married_dummy'] = np.where(df['civstatus'] == 'never-married',1.0,0.0)


#Para variable 'collars'

df['collars'] = df['collars'].replace(


['?',  'Prof-specialty','Exec-managerial','Adm-clerical','Sales',       'Tech-support','Craft-repair','Machine-op-inspct','Transport-moving','Handlers-cleaners','Farming-fishing','Protective-serv','Priv-house-serv','Other-service','Armed-Forces'],
 
[NA,'white-collar',  'white-collar',   'white-collar','white-collar','white-collar','blue-collar' ,'blue-collar',     'blue-collar',     'blue-collar',      'blue-collar',   'blue-collar',    'blue-collar',    'others',        'others'])

# Se crean variables dummy para 'collars'


df['white-collar_dummy'] = np.where(df['collars'] == 'white-collar',1.0,0.0)
df['blue-collar_dummy'] = np.where(df['collars'] == 'blue-collar',1.0,0.0)
df['others_dummy'] = np.where(df['collars'] == 'others',1.0,0.0)


#Se ordenan los países según lo requerido.

df['region'] = df['region'].replace('?',NA)
df['region'] = df['region'].replace(

['Peru','Ecuador','Columbia','Cuba','Puerto-Rico','Dominican-Republic','Haiti','Guatemala','El-Salvador','Nicaragua',
 'United-States','Mexico','Canada','Jamaica','Outlying-US(Guam-USVI-etc)','Honduras'], 'america')


df['region'] = df['region'].replace(

['Germany','England','Italy','Portugal','Poland','Greece','France','Ireland','Yugoslavia','Scotland','Hungary',
 'Holand-Netherlands'],'europa')
 
df['region'] = df['region'].replace(
 
['China','Japan','Taiwan','India','Vietnam','Iran','Hong','Thailand','Philippines','Laos'],'asia')


df['region'] = df['region'].replace(
    
    ['South','Trinadad&Tobago','Cambodia'],'africa')

#Se transforma variable income a dummy de forma tal que cuando el monto sea mayor a 50K se obtenga 1 y cuando sea menor 0. 

df['income'] = df['income'].replace(
    
    ['<=50K','>50K','?'],
    [0,1,NA])
```
