## Modelación predictiva

```python
# Modelo predictivo saturado
columnas = df.columns.tolist()
# Removemos de la lista el nombre del vector objetivo
for var in ['income', 'bin_income']:
    if var in columnas:
        columnas.remove(var)

# dividimos la muestra en entrenamiento y validación
x_train, x_test, y_train, y_test = train_test_split(StandardScaler().fit_transform(df.loc[:,columnas]), df['bin_income'], test_size = .33, random_state = 11238)
modelo_saturado = linear_model.LogisticRegression().fit(x_train, y_train)
```

```python
y_hat = modelo_saturado.predict(x_test)
confusion_matrix(y_test, y_hat)
```

Se obtuvieron:
- 10.386 verdaderos positivos
- 804 falsos positivos
- 2.163 verdaderos negativos
- 1.571 falsos negativos

```python
print(classification_report(y_test, y_hat))
```

Precision:
- la precisión de los verdaderos positivos sobre todos los positivos es de 87% 
- la precisión de los verdaderos negativos sobre todos los negativos es de 73%

Recall:
- El porcentaje de los verdaderos positivos sobre todos los positivos del vector entrenamiento es de 93%.
- El porcentaje de los verdaderos positivos sobre todos los positivos del vector entrenamiento es de 58%.

f1-score: $\dfrac{2*precision*recal}{precision+recal}$
- La tasa del puntaje f1 para los valores positivos es de 90%
- La tasa del puntaje f1 para los valores negativos es de 66%

```python
false_positive, true_positive, threshold = roc_curve(y_test, modelo_saturado.predict_proba(x_test)[:,1])

plt.plot(false_positive, true_positive)
plt.plot([0, 1], linestyle = "--")
plt.axhline(roc_auc_score(y_test, modelo_saturado.predict_proba(x_test)[:,1]), color = 'r')
plt.ylabel ("Verdaderos positivos")
plt.xlabel ("Falsos positivos")
print(roc_auc_score(y_test, modelo_saturado.predict_proba(x_test)[:,1]))
```

El ratio entre los verdaderos positivos y los verdaderos negativos es de 89,62% para el modelo saturado.
